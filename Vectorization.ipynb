```python
# Function to extract text and chunk it into sentences
def extract_and_chunk_pdf(pdf_file_path, chunk_size=5):
    all_text = ""

    # Step 1: Extract text from the PDF
    with pdfplumber.open(pdf_file_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            all_text += text

    # Step 2: Split the text into sentences
    sentences = re.split(r'(?<=[.!?]) +', all_text)  # Regex to split by sentence

    # Step 3: Chunk the sentences
    chunks = []
    for i in range(0, len(sentences), chunk_size):
        chunk = " ".join(sentences[i:i+chunk_size])
        chunks.append(chunk)

    return chunks

# Vectorization function using a pre-trained model
def vectorize_chunks(chunks):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # Pre-trained Sentence-BERT model
    vectors = model.encode(chunks)  # Encode each chunk into a vector
    return vectors

# Example usage
pdf_file_path = '/content/mod1.pdf'  # Replace with your PDF file path
chunks = extract_and_chunk_pdf(pdf_file_path, chunk_size=5)

# Vectorize the chunks
chunk_vectors = vectorize_chunks(chunks)

# Display chunk vectors
for idx, vector in enumerate(chunk_vectors, 1):
    print(f"Vector for Chunk {idx}:\n{vector}\n")
```

```python


# Initialize Pinecone connection
pc = Pinecone(api_key="b1938a13-f1f4-4498-b0ec-3765df4cd242")
index_name = "quickstart"

# Initialize SentenceTransformer model - Use the same model as for indexing
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # Same model for indexing and querying

# Function to calculate Euclidean distance
def euclidean_distance(vector1, vector2):
    return np.linalg.norm(np.array(vector1) - np.array(vector2))

# Function to calculate Cosine similarity
def cosine_similarity(vector1, vector2):
    vector1 = np.array(vector1)
    vector2 = np.array(vector2)
    cos_sim = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))
    return cos_sim

# Function to query Pinecone and compute distances
def search_pinecone(user_query, index_name="quickstart", top_k=5):
    # Get the Pinecone index
    index = pc.Index(index_name)  # Get the Pinecone index using the index_name

    # Convert the user query into a vector
    query_vector = model.encode(user_query).tolist()

    # Query Pinecone index for similar vectors
    result = index.query(vector=query_vector, top_k=top_k, include_metadata=True, include_values=True)

    # Display results with Euclidean distance and Cosine similarity
    print("Search results:")
    for match in result['matches']:
        match_id = match['id']
        score = match['score']
        chunk_text = match['metadata'].get('text', 'No text available')  # Retrieve text from metadata
        result_vector = match['values']  # Fetch the stored vector from Pinecone

        # Calculate Euclidean distance and Cosine similarity
        euclidean_dist = euclidean_distance(query_vector, result_vector)
        cos_sim = cosine_similarity(query_vector, result_vector)

        print(f"ID: {match_id}, Score: {score}")
        print(f"Text: {chunk_text}")
        print(f"Euclidean Distance: {euclidean_dist:.4f}")
        print(f"Cosine Similarity: {cos_sim:.4f}\n")

# Example usage
user_query = input("Enter your search query: ")
search_pinecone(user_query, top_k=5)
```
